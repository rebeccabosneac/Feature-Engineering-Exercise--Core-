```python
# Imports
import numpy as np
import pandas as pd
```

```python
# import the data
df = pd.read_csv(r"C:\Users\romo_\Documents\GitHub\Feature-Engineering-Exercise--Core-\bikeshare_train - bikeshare_train.csv")
df.head()
```

```python
# drop the 'casual' and 'registered' columns
df.drop(columns=['casual','registered'], inplace=True)
df.head()
```

```python
df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.info()
```

```python
# transforming datetime column
df['month (name)'] = df['datetime'].dt.month_name()
df['day of week (name)'] = df['datetime'].dt.day_name()
df['hour of day'] = df['datetime'].dt.time
df.head()
```

```python
# drop the 'datetime' and 'season' columns
df.drop(columns=['datetime','season'], inplace=True)
df.head()
```

```python
# converting temp from Celsius to Fahrenheit
df['temp'] = df.apply(lambda x: (9/5)*x['temp']+32,axis=1)
df['atemp'] = df.apply(lambda x: (9/5)*x['atemp']+32,axis=1)
df.head()
```

```python
df['temp_variance'] = df['temp'] - df['atemp']
df.head()
```

```python
avg_temp = df['atemp'].mean()

def bin_temp(temp):
    if temp < 0 :
        return 'Colder'
    else:
        return 'Warmer'
```

```python
# create a new column, 'temp_variance,
df['temp_variance'] = df['temp_variance'].apply(bin_temp)
df.head()
```

```python
# drop the 'atemp' column
df.drop('atemp', inplace=True, axis=1)
df.head()
```
